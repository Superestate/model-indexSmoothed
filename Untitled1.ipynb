```python
import sys
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
   
```

```python
data=pd.read_csv("rand_Sales.csv")
data.head(5)
```

```python
data['Date'] =pd.to_datetime(data.Date)
```

```python
data=data.sort_values('Date')
data['Month']=data['Date'].dt.month
data['Year']=data['Date'].dt.year
```

```python
data.head()
```

```python
filtered=data.groupby('ID').filter(lambda x: len(x) > 1)
```

```python
filtered.head(5)
```

```python
filtered.set_index(['Month', 'Postcode'], inplace=True)
filtered


```

```python
#median=filtered.groupby(filtered.Month)[['Price']].median()
grouped = filtered.groupby(level = ['Month','Postcode'])
filtered['Median'] = grouped['Price'].transform('median')
filtered
```

```python
filtered=filtered.reset_index()
```

```python
dummies=pd.get_dummies(filtered['Date'])

filtered = pd.concat([filtered, dummies], axis=1)
```

```python
filtered.set_index('Month', inplace=True)
```

```python
p=100/filtered.iloc[0,11]
filtered.iloc[:,11]*=p
filtered
```

```python
#filtered=filtered.reset_index()
filtered.plot(x ='Month', y='Median', kind = 'line')
plt.show()
```
